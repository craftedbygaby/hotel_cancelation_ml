# 02. Importing Data

```python
data = pd.read_csv('../01. Data/hotel_bookings_raw_data.csv')
```

# 03. Cleaning Data

```python
data.isnull().sum()
```

```python
# dropping agent and company columns due to number os nulls
data = data.drop(columns=['agent', 'company'])
```

```python
# dropping empty rows in colums that have them
data = data.dropna(subset=['children', 'country'])
```

```python
data.isnull().sum()
```

```python
# checking for duplicated entries
data.duplicated().sum()
```

```python
# dropping duplicates
data = data.drop_duplicates()
```
